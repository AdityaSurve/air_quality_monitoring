```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
import csv

webdriver_service = Service(r'C://Users/Aditya/Downloads/msedgedriver.exe')

driver = webdriver.Edge(service=webdriver_service)


def write_to_csv(data):
    csv_file_path = 'output.csv'
    csv_header = ["DateObserved", "HourObserved", "LocalTimeZone", "ReportingArea", "StateCode",
                  "Latitude", "Longitude", "ParameterName", "AQI", "CategoryNumber", "CategoryName"]
    file_exists = os.path.isfile(csv_file_path)
    with open(csv_file_path, 'a', newline='') as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=csv_header)
        if not file_exists:
            writer.writeheader()
        writer.writerows(data)

try:
    driver.get('https://docs.airnowapi.org/login')

    username_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'username'))
    )
    password_field = driver.find_element(By.ID, 'password')
    login_button = driver.find_element(By.NAME, 'exec')

    username_field.send_keys('aditya_25')
    password_field.send_keys('aditya_25@AirNow')

    login_button.click()

    time.sleep(5)

    driver.get('https://docs.airnowapi.org/HistoricalObservationsByLatLon/query')

    input1 = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'latitudeCoord'))
    )
    input2 = driver.find_element(By.ID, 'longitudeCoord')
    input3 = driver.find_element(By.NAME, 'distance')
    button = driver.find_element(By.ID, 'buildUrl')

    input1.clear()
    input2.clear()
    input3.clear()
    input1.send_keys('37.7749')  # San Francisco, California
    input2.send_keys('-122.4194')  # San Francisco, California
    input3.send_keys('15')

    button.click()

    output = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'urlView'))
    )

    output_text = output.text
    time.sleep(1)

    main_button = driver.find_element(By.ID, 'runItBtn')

    main_button.click()

    final_output = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'outputView'))
    )

    time.sleep(5)
    print(final_output.text)
    print(output_text)

    output_data = []
    for line in final_output.text.split('\n')[1:]:
        line_data = line.split(',')
        output_data.append({
            "DateObserved": line_data[0].replace('"', ''),
            "HourObserved": line_data[1].replace('"', ''),
            "LocalTimeZone": line_data[2].replace('"', ''),
            "ReportingArea": line_data[3].replace('"', ''),
            "StateCode": line_data[4].replace('"', ''),
            "Latitude": line_data[5].replace('"', ''),
            "Longitude": line_data[6].replace('"', ''),
            "ParameterName": line_data[7].replace('"', ''),
            "AQI": line_data[8].replace('"', ''),
            "CategoryNumber": line_data[9].replace('"', ''),
            "CategoryName": line_data[10].replace('"', ''),
        })

    write_to_csv(output_data)

finally:
    driver.quit()
```
